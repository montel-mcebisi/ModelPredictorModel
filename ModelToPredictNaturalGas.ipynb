# Libraries

```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Uploading file

```python
from google.colab import drive
drive.mount('/content/drive')

df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/Nat_Gas.csv')
df.head()
```

# Code

```python
# Clean and convert dates
df['Date'] = pd.to_datetime(df['Dates'])
df = df.sort_values('Date')

# Convert to ordinal for regression
df['DateOrdinal'] = df['Date'].map(datetime.toordinal)
```

# Fit Linear Regression Model

```python
# Model
model = LinearRegression()
X = df[['DateOrdinal']]
y = df['Prices']
model.fit(X,y)
```

# Prediction Function

```python
def predict_price(date_string):
  try:
    input_date = datetime.strptime(date_string, "%Y-%m-%d")
    input_ordinal = np.array([[input_date.toordinal()]])
    predicted_price = model.predict(input_ordinal)[0]
    return round(predicted_price,2)
  except ValueError:
    return "Invalid date format. input YYYY-mm-dd"


user_input = input("Enter a date(YYYY-MM-DD):")
predicted_price = predict_price(user_input)
print(f"estimated natural gas price on {user_input}: ${predicted_price}")
```
